```
#PEGANDO OS DADOS DAS INFORMAÇÕES DOS FUNDOS E FILTRANDO SOMENTE OS MULTIMERCADOS

import pandas as pd
import numpy
import requests 

# URL do arquivo

url = 'https://dados.cvm.gov.br/dados/FI/DOC/EXTRATO/DADOS/extrato_fi.csv'

# lê o csv, define o separador e especifica a codificação

df = pd.read_csv(url, sep=';', encoding='ISO-8859-1')

# Puxa somente as 3 colunas

df = df.loc[:, ['CNPJ_FUNDO', 'DENOM_SOCIAL', 'CLASSE_ANBIMA']]

# Tira os NA da coluna Classe Anbima

df['CLASSE_ANBIMA'] = df['CLASSE_ANBIMA'].fillna('')

# Puxa somente os resultados que tem multimercado escrito

df = df.loc[df['CLASSE_ANBIMA'].str.contains('MULTIMERCADO')]

df
```

```
# PEGANDO OS DADOS DE PATRIMÔNIO LÍQUIDO DOS FUNDOS

import pandas as pd
import urllib.request
import zipfile

# URL do arquivo ZIP
url = 'https://dados.cvm.gov.br/dados/FI/DOC/INF_DIARIO/DADOS/inf_diario_fi_202203.zip'

# Baixa o arquivo ZIP
filename, headers = urllib.request.urlretrieve(url)

# Extrai o arquivo CSV
with zipfile.ZipFile(filename, 'r') as zip_ref:
    zip_ref.extractall()

# Carrega o arquivo CSV
df2 = pd.read_csv('inf_diario_fi_202203.csv', sep=';', decimal=',', usecols=['CNPJ_FUNDO', 'VL_TOTAL'])

# Imprime as primeiras linhas do DataFrame
df2
```

```
# CONECTANDO A PRIMEIRA TABELA COM A SEGUNDA
df_merged = pd.merge(df, df2, on='CNPJ_FUNDO')
df_merged = df_merged[['CNPJ_FUNDO', 'DENOM_SOCIAL', 'CLASSE_ANBIMA', 'VL_TOTAL']]
df_merged.columns = ['CNPJ', 'DENOM_SOCIAL', 'CLASSE_ANBIMA', 'VL_TOTAL']
```

```
#FILTRANDO DO MAIOR PARA O MENOR 

df_merged = df_merged.sort_values(by='VL_TOTAL', ascending=False)
df_merged
```
